```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=sns.load_dataset('iris')
df.shape

df.columns
df.head()
df.info()
df.describe()

sns.scatterplot(data=df,x='sepal_length',y='sepal_width')
plt.show()

df.hist(figsize=(10,10))
plt.show()

sns.boxenplot(data=df)
plt.show()
```

```python
# Step1: Hugging Face pipeline import kar rahe hain
from transformers import pipeline

#Pre-trained AI model load ho raha hai
#GPT2 is basic language model hai jo generate karta hai
chatbot = pipeline("text-generation", model="gpt2")
```

```python

#Step2:AI response function
def get_health_response(user_question):

    prompt = f"""
You are a helpful medical assistant.
Give simple and safe health advice.

Question: {user_question}
Answer:
"""
#Step3:AI model reponse generate karna
    result = chatbot(prompt, max_length=100, num_return_sequences=1)

#step4:AI ka generated text return karna    
    return result[0]["generated_text"]
```

```python

#step4:Chatbot start message
print("Health Chatbot (hugging face free version)")
print("Type exist to stop\n")

# Step5:infinite loop 
while True:

    
    user_input = input("Ask health question (or type exit)")

    if user_input.lower()=="exit":
        break
        
    answer = get_health_response(user_input)
    print("Bot:",answer)
```

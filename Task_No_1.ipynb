```python

# step1: To import the libaries
import pandas as pd  # Handling data
import matplotlib.pyplot as plt # For graphs
import seaborn as sns # To load datasheet

#step2:Store value in varible df iris is datadsheet
df=sns.load_dataset('iris')
df.shape
df.columns
df.head()
df.info()
df.describe()

#step3:extension of sns for scatterplot
sns.scatterplot(data=df,x='sepal_length',y='sepal_width')
plt.show()

#step4:extension of df for histogram
df.hist(figsize=(10,10))
plt.show()

#step5:extension of sns for boxenplot
sns.boxenplot(data=df)
plt.show()
```

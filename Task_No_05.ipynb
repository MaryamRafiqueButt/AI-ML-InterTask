```python
#Step 2: Hugging face pipeline import kar raha hai
from transformers import pipeline
```

```python
# GPT2 free model load ho raha hai
chatbot = pipeline("text-generation", model="gpt2")
```

```python
#Step 3:Safety functions
def safety_check(question):
    dangerous_words = ["overdose", "suicide", "self harm"]

    for word in dangerous_words:
        if word in question.lower():
            return False # unsafe question
    return True # safe question
```

```python
# Step 5:AI response function
def get_health_response(user_question):


# Step 6:AI ka samjhny ky liya prompt banya hai
    prompt = f"""
You are a helpful medical assistant.
Give simple, safe and general health advice.

Question: {user_question}
Answer:
"""
#Step 7:chatbox text generate karna
    result = chatbot(prompt, max_length=100, num_return_sequences=1)

#Step 8:AI ka generated text return karna
    return result[0]["generated_text"]
```

```python
#Step 9:Chatbox start message
print(" Health Chatbot Started (Google Colab)")
print("Type 'exit' to stop\n")


#Step 10:infinite looop (ask again again questions )
while True:


    user_input = input("Ask health question: ")

    if user_input.lower() == "exit":
        print("Goodbye ")
        break

    if not safety_check(user_input):
        print("Sorry, unsafe question. Please consult a doctor.")
        continue

    answer = get_health_response(user_input)
    print("\nBot:", answer, "\n")
```
